```python
from src.load_data.load_data import load_data

path = '../data/raw/airbnb.csv'

raw_df = load_data(path)
```

```python
raw_df.drop(columns=['Unnamed: 0', 'id', 'name', 'host_name', 'host_id', 'features', 'safety_rules', 'hourse_rules', 'img_links', 'checkin', 'checkout'], inplace=True)
```

```python
raw_df
```

```python
unique_address = raw_df["address"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
unique_address
```

```python
geocode_results = {}
```

```python
from src.utils.open_cage_geocode import get_coordinates
import time

```

```python
for address in unique_address["address"]:
    try:
        lat, lng = get_coordinates(address)  # Zavolanie tvojej funkcie
        geocode_results[address] = (lat, lng)
    except ValueError as e:
        # Ak funkcia zlyhá, pridaj None a pokračuj
        print(e)
        geocode_results[address] = (None, None)

    # Obmedzenie rýchlosti API volania (napr. 2 požiadavky za sekundu)
    time.sleep(0.5)
```

```python
# Zobrazenie výsledného datasetu
print(geocode_results)
```

```python
raw_df['latitude'] = raw_df['address'].map(lambda addr: geocode_results.get(addr, (None, None))[0])
raw_df['longitude'] = raw_df['address'].map(lambda addr: geocode_results.get(addr, (None, None))[1])

# Výsledný dataset
raw_df.head()
```

```python
raw_df['latitude'].isna().sum()
```

```python
raw_df['longitude'].isna().sum()
```

```python
raw_df.dropna(subset=['latitude', 'longitude'], inplace=True)
```

```python
raw_df.info()
```

```python
raw_df.drop(columns=["address"], inplace=True)
```

```python
raw_df['rating'].replace("New", 0)
```

```python
raw_df['rating'] = raw_df['rating'].replace("New", 0).astype(float).fillna(0)
```

```python
raw_df['reviews'] = raw_df['reviews'].str.replace(',', '').astype(float)
```
